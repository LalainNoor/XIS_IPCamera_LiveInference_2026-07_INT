# IP Camera Live Inference

## Objective

Develop a live IP camera inference pipeline capable of receiving frames from an IP camera (RTSP/HTTP), performing real-time image classification using ONNX Runtime, and displaying predictions together with FPS, latency, and throughput.

The application also supports USB webcams and prerecorded videos for testing while maintaining a unified inference pipeline.

## Project Components

- Camera Module (`camera.py`)
- Configuration (`config.py`)
- Frame Preprocessing (`preprocess.py`)
- ONNX Runtime Inference (`inference.py`)
- Performance Metrics (`metrics.py`)
- Live Display (`display.py`)
- Main Application (`main.py`)

### Features

- IP Camera (RTSP/HTTP) support
- USB Webcam support
- Video file support
- Automatic reconnection handling
- Live prediction overlay
- FPS monitoring
- Latency measurement
- Throughput calculation
